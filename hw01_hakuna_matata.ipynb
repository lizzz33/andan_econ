### Задача 1 (Парадокс инспектора)

Автобусы отходят от автостанции с 8:00 до 20:00. Первый автобус отходит ровно в 8:00. Затем интервалы между автобусами случайны, независимы и равновероятно равны либо 5-и, либо 10-и минутам. Будем считать, что за минуту на автостанцию приходит ровно один пассажир, и все пассажиры едут ближайшим автобусом.

Проведите $10^4$ экспериментов и с их помощью:

__а) [5 баллов]__ Постройте гистограмму количества автобусов, отошедших от автостанции за сутки. Похоже ли визуально распределение на нормальное?

__б) [10 баллов]__ Инспектор Тимон выбирает равновероятно один из всех автобусов отошедших от автостанции за сутки. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__в) [10 баллов]__ Инспектор Пумба приходит на автостанцию в случайный момент времени, равномерный от 8:00 до 20:00 и садится в первый пришедший автобус. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__г) [10 баллов]__ Как изменятся ответы на эти вопросы, если время между автобусами будет экспоненциально со средним в 10 минут?

__Пояснения:__ первый автобус можно не включать в выборку, Пумбу за пассажира можно не считать.

```python
import numpy as np
import pandas as pd
from scipy import stats

# библиотеки для визуализации
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('ggplot')

np.random.seed(123)
```

```python
time = []
people = []
for i in range(10000):
    bus = np.zeros(int(60/5*12+1))
    queue = np.zeros(int(60/5*12+1))
    bus[0] = 1
    for i in range(1, len(bus)):
        if bus[i-1] == 0:
            bus[i] = 1
            queue[i] = 10
        else:
            bus[i] = np.random.choice([0,1], size=1)[0]
            queue[i] = 5
    time.append(bus)
    people.append(queue)
```

```python
sum_ = np.sum(time, axis=1)
plt.hist(sum_, bins=14, color='g');
plt.title('Количество автобусов, отошедших от автостанции за сутки')
plt.xlabel('Кол-во автобусов')
plt.ylabel('Кол-во случаев')
plt.show()
```

На графике приведена гистограмма кол-ва автобусов, отошедших от станции за сутки. Большая вероятность, что будет 95-100 автобусов. Разброс от 85 до 110 автобусов. Визуально распередение похоже на нормальное.

```python
people_ = (np.array(people) * np.array(time))
```

```python
Timon = list(people_.copy())
for i in range(len(Timon)):
    for j in range(list(Timon[i]).count(0)):
        Timon[i] = np.append(np.delete(Timon[i], list(Timon[i]).index(0)), 0)
```

```python
k = np.random.choice(np.arange(1, len(Timon[0])))
plt.hist(np.array(Timon)[:,k], color='magenta');
plt.title('Количество пассажиров в автобусе для Тимона')
plt.xlabel('Кол-во пассажиров')
plt.ylabel('Кол-во случаев')
plt.show()
```

```python
print('Мат. ожидание ', np.mean(np.array(Timon)[:,k]))
print('Дисперсия ', np.var(np.array(Timon)[:,k]))
```

Тимон выбрал случайный автобус из всех отошедших от станции за сутки. График демонстирирует почти равновероятный исход кол-ва пассажиров в автобусе для Тимона - либо 5, либо 10 человек.

```python
Pumbaa = (np.array(people_)[:,k]).copy()
Pumbaa[Pumbaa == 0] = 10
```

```python
plt.hist(Pumbaa, color='aquamarine');
plt.title('Количество пассажиров в автобусе для Пумбы')
plt.xlabel('Кол-во пассажиров')
plt.ylabel('Кол-во случаев')
plt.show()
```

```python
print('Мат. ожидание ', np.mean(Pumbaa))
print('Дисперсия ', np.var(Pumbaa))
```

Пумба выбрал пришел на автобус в случайный момент в рабочее для автобусов время. Для него вероятность попась в автобус с 10 пассажирами в два раза больше, чем попасть в автобус с 5 пассажирами.

Важное напоминание: каждую минуту на остановку подходит новый человек, поэтому число пассажиров можно интерпретировать как время ожидания автобуса.


В этом и проявляется парадокс инспектора: случайное событие (приход Пумбы к автобусу в случайный момент времени) с большей вероятностью попадет в длинный интервал (большее кол-во пассажиров).

Плотность вероятности экспоненциального распределения, описывающего время между автобусами:

$f(t) = 10*exp(-10t)$

```python
exp_rv = stats.expon(loc=10)
exp_time = []
for i in range(10000):
    exp_time.append(np.floor(exp_rv.rvs(12*6)))
```

```python
k = np.random.choice(np.arange(1, 12*6))
plt.hist(np.array(exp_time)[:,k], color='magenta');
plt.title('Количество пассажиров в автобусе для Тимона')
plt.xlabel('Кол-во пассажиров')
plt.ylabel('Кол-во случаев')
plt.show()
```

```python
print('Мат. ожидание ', np.mean(exp_time))
print('Дисперсия ', np.var(exp_time))
```

В случае распределения времени между автобусами по экспоненциальному закону с мат. ожиданием 10 мин, кол-во пассажиров в автобусе будет не менее 10 человек.

```python
k = np.random.choice(np.arange(1, 12*60))
exp_pumb = []
for i in range(len(exp_time)):
    for j in range(len(exp_time)):
        if sum(exp_time[i][:j]) >= k:
            exp_pumb.append(exp_time[i][j-1])
            break
```

```python
plt.hist(np.array(exp_pumb), color='aquamarine');
plt.title('Количество пассажиров в автобусе для Пумбы')
plt.xlabel('Кол-во пассажиров')
plt.ylabel('Кол-во случаев')
plt.show()
```

```python
print('Мат. ожидание ', np.mean(exp_pumb))
print('Дисперсия ', np.var(exp_pumb))
```

Вновь можно наблюдать, что автобус в случайно выбранное время с большей вероятностью окажется более загруженным. Однако, следует заметить, что разница между наблюдениями Тимона и Пумбы в случае экспоненциального распределения довольно мала.

### Задача 2 (Парадокс Хуана Мануэля Родригеса Паррондо)

У Атоса, Портоса и Арамиса по 1000 франков. Атос постоянно ходит в казино А, где каждый раз выигрывает один франк с вероятностью $0.49$ и проигрывает один франк с вероятностью $0.51$. Портос ходит в казино Б, где ситуация интереснее :)

- Если богатство посетителя кратно трём, то посетитель выигрывает франк с вероятностью $0.09$ и проигрывает один франк с вероятностью $0.91$
- Если богатство посетителя не кратно трём, то посетитель выигрывает франк с вероятностью $0.74$ и проигрывает один франк с вероятностью $0.26$

Арамис каждый раз выбирает казино А или казино Б равновероятно. Проведите $10^4$ симуляций эволюции благосостояния Атоса, Портоса и Арамиса.Каждая симуляция предусматривает $1000$ посещений казино. Все три игрока все деньги носят с собой и ни на что не тратят :)

__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__б) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Портоса, по горизонтали —  номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__в) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Арамиса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.
    

```python
athos = []
for i in range(10000):
    cash = list(np.random.choice([1, -1], size=1000, p=[0.49, 0.51]))
    cash.insert(0, 1000)
    for j in range(1, len(cash)):
        cash[j] += cash[j-1]
    athos.append(cash)
```

```python
from matplotlib.pyplot import figure

figure(figsize=(25, 5), dpi=80)
for i in range(10):
    k = np.random.choice(np.arange(0, 10000))
    plt.plot(athos[k])
plt.plot(np.mean(athos, axis=0), label='Усредненная траектория')
plt.xlim([0, 1000])
plt.legend()
plt.title('Изменения богатства Атоса')
plt.xlabel('Номер похода в казино')
plt.ylabel('Богатство')
plt.show()
```

На графике показаны траектории изменения богатства Атоса. Есть как прибыльные случаи, так и сильно проигрышные. В среднем, Атос теряет 20 франков.

```python
porthos = []
for i in range(10000):
    cash = np.zeros(1001)
    cash[0] = 1000
    for j in range(1, len(cash)):
        if cash[j-1] % 3 == 0:
            cash[j] = cash[j-1] + np.random.choice([1, -1], size=1, p=[0.09, 0.91])
        else:
            cash[j] = cash[j-1] + np.random.choice([1, -1], size=1, p=[0.74, 0.26])
    porthos.append(cash)
```

```python
figure(figsize=(25, 5), dpi=80)
for i in range(10):
    k = np.random.choice(np.arange(0, 10000))
    plt.plot(porthos[k])
plt.plot(np.mean(porthos, axis=0), label='Усредненная траектория')
plt.xlim([0, 1000])
plt.legend()
plt.title('Изменения богатства Портоса')
plt.xlabel('Номер похода в казино')
plt.ylabel('Богатство')
plt.show()
```

На графике показаны траектории изменения богатства Портоса. Заметно, что удачливость Пороса немного больше. 10 случайно выбранных траекторий менее разбросаны, то есть колебания в богатстве меньше, чем у Атоса. В среднем, Портоса теряет 15 франков.

```python
aramis = []
for i in range(10000):
    cash = np.zeros(1001)
    cash[0] = 1000
    for j in range(1, len(cash)):
        if np.random.choice(['a', 'b'], size=1, p=[0.5, 0.5]) == 'a':
            cash[j] = cash[j-1] + np.random.choice([1, -1], size=1, p=[0.49, 0.51])
        else:
            if cash[j-1] % 3 == 0:
                cash[j] = cash[j-1] + np.random.choice([1, -1], size=1, p=[0.09, 0.91])
            else:
                cash[j] = cash[j-1] + np.random.choice([1, -1], size=1, p=[0.74, 0.26])
    aramis.append(cash)
```

```python
figure(figsize=(25, 5), dpi=80)
for i in range(10):
    k = np.random.choice(np.arange(0, 10000))
    plt.plot(aramis[k])
plt.plot(np.mean(aramis, axis=0), label='Усредненная траектория')
plt.xlim([0, 1000])
plt.legend()
plt.title('Изменения богатства Арамиса')
plt.xlabel('Номер похода в казино')
plt.ylabel('Богатство')
plt.show()
```

На графике наблюдаем исходы стратегии Арамиса. Объединив два неудачных подхода к походам в казино, он получает довольно прибыльную траекторию. В среднем Арамис выигрывает 5 франков.

Задача демонстрирует парадокс Паррондо. Применяя равновероятно проигрышные стратергии, получется выигрышная.  

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются
нормальным распределением с ожиданием 60 и стандартным отклонением 10.

УШЭ (Урюпинская Школа Экономики) ныне престижна и забирает себе всех школьников набравших более $n$ баллов в сумме по русскому и математике.

Для каждого $n$ от $80$ до $160$ с шагом $5$ случайно создайте $10^4$ школьников и поделите их на прошедших и не прошедших в УШЭ.


__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$.

__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

```python
norm_rv = stats.norm(loc=60, scale=10)
df_res = pd.DataFrame({'rus': np.round(norm_rv.rvs(10000)), 'math': np.round(norm_rv.rvs(10000))})
df_res['sum'] = df_res['rus'] + df_res['math']
df_res.head()
```

```python
n = np.arange(80, 165, 5)
count_n = []
for i in range(len(n)):
    count_n.append(sum(df_res['sum'] > n[i]))
```

```python
plt.plot(n, count_n)
plt.title('Число прошедших в УШЭ в зависимости от n')
plt.xlabel('n')
plt.ylabel('Кол-во прошедших')
plt.show()
```

Зависимость числа прошедших от проходного балла описывается функцией, похожей на обратную сигмоиду с верхней границей равной размеру выборки.

```python
cov_pos = []
for i in range(len(n)):
    cov_pos.append(df_res[df_res['sum']>n[i]]['rus'].corr(df_res[df_res['sum']>n[i]]['math']))

plt.plot(n, cov_pos)
plt.title('Корреляция результатов по рус. и мат. среди прошедших в УШЭ')
plt.xlabel('n')
plt.ylabel('Корреляция')
plt.show()
```

Среди прошедших наблюдается увеличение обратной зависимости между баллами по русскому и по математике от суммы баллов. Так, чем больше у ученика сумма баллов, тем больше отрицательная корреляция между результатами двух экзаменов. Тогда как при низких баллах зависимости почти нет.

```python
cov_neg = []
for i in range(len(n)):
    cov_neg.append(df_res[df_res['sum']<=n[i]]['rus'].corr(df_res[df_res['sum']<=n[i]]['math']))

plt.plot(n, cov_neg)
plt.title('Корреляция результатов по рус. и мат. среди НЕ прошедших в УШЭ')
plt.xlabel('n')
plt.ylabel('Корреляция')
plt.show()
```

Среди непрошедших обратная ситуация. Чем больше сумма баллов, тем меньше отрицательная корреляция баллов между двумя предметами. Чем меньше баллов у ученика в сумме, тем больше разница баллов по русскому и математике.

Таким образом, пример демонстрирует парадокс Берксона. Наблюдаем отрицательную корреляцию между двумя положительными характеристиками. Логично было бы, если с увеличением суммы баллов, баллы по каждому из предметов увеличивались, но график показал обратное. И в другом случае, ожидалось, что чем хуже результат, тем меньше обе оценки, но график показал, что отрицательную корреляцию между двумя отрицательными характеристиками.
